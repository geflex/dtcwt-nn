```python
import matplotlib.mlab as mlab

# _ = plt.gca().specgram(sample.sample, Fs=sample.fs)
fig, ax = plt.subplots(3, 3, figsize=(14, 8))
ax = ax.flatten()

mspec, mfreqs = mlab.magnitude_spectrum(sample.sample)
ax[0].set_title('Magnitude spectrum')
ax[0].plot(mfreqs, mspec)

ax[1].set_title('FFT')
ax[1].plot(*sample.spectre())

ax[2].set_title('Specgram')
# spec, freqs, t = mlab.specgram(sample.sample)
# ax[2].imshow(spec, aspect='auto', origin='upper', extent=(np.min(t), np.max(t), freqs[0], freqs[-1]))
ax[2].specgram(sample.sample)

ax[3].set_title('DTCWT')
ax[3].imshow(np.abs(sample.dtcwt_scalogram()), aspect='auto')

ax[4].set_title('CWT morlet')
ax[4].imshow(np.abs(sample.cwt_morlet(w=3.0)), aspect='auto')

ax[5].set_title('CWT ricker')
ax[5].imshow(sample.cwt_ricker(), aspect='auto')

fig.tight_layout()
```
